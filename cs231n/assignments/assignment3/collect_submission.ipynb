```
# Local setup: run this notebook from the repo root or from assignment3.
from pathlib import Path
import os
import sys


def find_assignment_dir():
    cwd = Path.cwd().resolve()
    for base in (cwd, *cwd.parents):
        if base.name == "assignment3" and (base / "collectSubmission.sh").exists():
            return base
        candidate = base / "cs231n" / "assignments" / "assignment3"
        if candidate.exists():
            return candidate.resolve()
    raise RuntimeError(
        "Could not find cs231n/assignments/assignment3. "
        "Start Jupyter from the repo root or the assignment3 directory."
    )


ASSIGNMENT_DIR = find_assignment_dir()
os.chdir(ASSIGNMENT_DIR)
if str(ASSIGNMENT_DIR) not in sys.path:
    sys.path.insert(0, str(ASSIGNMENT_DIR))

print(f"Running assignment from: {ASSIGNMENT_DIR}")
```
